```python
from pyAudioAnalysis import audioTrainTest as aT
from collections import defaultdict
import os
```

```python
with open('project/download/erkennung.txt','rb') as f:
    lines = f.readlines()
```

```python
yes=[]
no=[]
for line in lines[1:]:
    temp = line.decode('utf-8','ignore').strip('\r\n').split('\t')
    naturalness= float(temp[-1].replace(',','.'))
#     print(naturalness)
    if naturalness>50:
        yes.append(temp[1])
    else:
        no.append(temp[1])
```
